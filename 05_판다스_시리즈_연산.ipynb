```python
# 브로드캐스팅: 연산이 시리즈 객체의 전체 값에 적용 (반복문 필요X)

from pandas import Series

s = Series([100, 200, 300])
print(s+10)
```

```python
# 사칙연산1: 동일한 인덱스를 갖는 값끼리 사칙연산이 적용됨

from pandas import Series

high = Series([51500, 51200, 52500, 51500, 51000])
low = Series([50700, 49000, 30000, 50800, 40600])

diff = high - low

print(diff)
```

```python
# 사칙연산2: 동일한 인덱스를 갖는 값끼리 사칙연산이 적용됨
# 순서와 상관없이 인덱스를 기준으로 계산
# 인덱스가 한 쪽의 시리즈 객체에만 존재하는 경우 사칙연산 불가

from pandas import Series

high1 = Series(data=[51500, 51200, 52500], index=['5/1', '5/2', '5/3'])
low1 = Series(data=[50700, 49000, 30000], index=['5/1', '5/5', '5/2'])

diff1 = high1 - low1
print(diff1)

# NaN: Not a Number
# 인덱스가 한 쪽에만 존재 -> NaN 발생
# NaN이 있을 경우 전체 dtype이 float으로 바뀜
```

```python
# 시리즈 비교 연산
# 비교 연산자의 결과로 Boolean 타입이 저장된 새로운 시리즈 객체가 리턴됨

from pandas import Series

s = Series([100, 200, 300, 400, 500])
cond = s > 300 # cond: Boolean 타입의 새 시리즈 객체
print(cond)
```

```python
# 시리즈 필터링1
# True/False 값을 통해 True 값만 필터링

from pandas import Series

s = Series([100, 200, 300, 400, 500])
cond = [False, True, False, False, True]
s[cond]
```

```python
# 시리즈 필터링2
# 비교 연산자의 결과 시리즈(ex: cond)를 사용하여 인덱싱 -> 조건을 만족하는(True) 데이터만 필터링

from pandas import Series

s = Series([100, 200, 300, 400, 500])
cond = s > 300 # cond = [False, False, False, True, True]

print(s[cond])
```

```python
# 시리즈 필터링3
# 시가가 3000원 이상인 날의 종가 출력

siga = Series([1000, 2000, 3000, 4000, 5000])
jongga = Series([1111, 2222, 3333, 4444, 5555])

condition = siga >= 3000 # condition = [False, False, True, True, True]

print(jongga[condition])
```

```python
# 연습문제1
# 고가와 저가 차이가 100 이상인 날의 고가 출력

from pandas import Series

low_price = Series([10, 200, 200, 400, 550])
high_price = Series([100, 300, 400, 500, 600])

cond = (high_price - low_price) >= 100

print(high_price[cond])
```

```python
# 연습문제2
# 종가가 80000원 이상 90000원 미만인 날짜 출력

jongga = Series([93000, 82400, 99100, 81000, 72300], ['5/14', '5/15', '5/16', '5/17', '5/18'])

# 파이썬 논리 연산자: and, or, not
# 판다스에서 이용 시: &, |, ~

cond = (jongga >= 80000) & (jongga < 90000)

# cond = (jongga >= 80000) and (jongga < 90000): and 사용시 오류

print(jongga[cond]) # 날짜와 종가 모두 출력

# 날짜만 출력하고 싶을 경우
jongga[cond].index

# 날짜만 반복문으로 출력하고 싶을 경우
for date in jongga[cond].index:
  print(date)
```
